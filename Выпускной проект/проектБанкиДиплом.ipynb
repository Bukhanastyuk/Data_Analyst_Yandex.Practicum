## Сегментация пользователей на основе потребления.

    Построим сводную таблицу и визуализацию доли ушедших клиентов в разрезе кол-ва продуктов.
    На тепловой карте также была слабо выраженная положительная корреляция оттока и таких параметров как имущество и активность. Рассмотрим отточнось в разрезе этих и других параметров.

```python
#  напишем цикл для построения сводных таблиц и графиков по соответствующим колонкам
column_list = ['products', 'equity', 'last_activity', 'credit_card', 'gender', 'city']

for column in column_list:
    pivot_table = df.pivot_table(index=column, values = 'churn', aggfunc = ['sum', 'count','mean']).reset_index()
    pivot_table['%'] = round(pivot_table['mean']*100, 2)
    pivot_table.columns = pivot_table.columns.droplevel(level=1)
    print(f'Доля ушедших клиентов по параметру {column}')
    display(pivot_table)
    
    # построим столбчатую диаграмму по сводной таблице
    f, ax = plt.subplots(figsize = (10,6))
    sns.set_color_codes('pastel')
    sns.barplot(x = column, y = 'count', data = pivot_table,
            label = 'Всего', color = 'b', edgecolor = 'w')
    sns.set_color_codes('muted')
    sns.barplot(x = column, y = 'sum', data = pivot_table,
            label = 'из них отточных', color = 'b', edgecolor = 'w')
    ax.legend(ncol = 2, loc = 'lower right')
    ax.set(ylabel="Количество клиентов", xlabel=f"Значение {column}")
    sns.despine(left = True, bottom = True)
    plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1))
    plt.title(f'Доля ушедших клиентов по параметру {column}')
    plt.title(column, fontsize = 16, color = 'black')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    # Добавление подписей процентов
    for i, percent in enumerate(pivot_table['%']):
        ax.text(i, pivot_table['%'].values[i] + 10, f'{percent}%', ha='center')
    plt.show()
```

## Формулирование и проверка гипотез.

### Проверьте гипотезу различия возраста между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним.

Для проверки гипотез о схожести/различии среднего возраста в двух сегментах, воспользуемся критерием Стьюдента с помощью двустороннего t-теста.

**Сформулируем гипотезы по следующим правилам:**
- Нулевую гипотезу H₀ сформулируем так, чтобы использовать знак равенства (неравенство НЕ предполагается), т.е. гипотеза учитывает отсутствие различий в выборках, отсутствие влияние каких-либо факторов.
- Альтернативная гипотеза формулируется исходя из того, что это утверждение, которое принимается верным, если отбрасывается H₀.

Гипотеза H0: Средний возраст клиентов, которые пользуются двумя продуктами банка и которые пользуются одним, одинаковый. 

Альтернативная гипотеза H1: Средний возраст клиентов, которые пользуются двумя продуктами банка и которые пользуются одним, различается. 

```python
# создадим выборки для клиентов с 1 и 2 продуктами
df_product_1 = df[df['products'] == 1]['age']
df_product_2 = df[df['products'] == 2]['age']
```

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

# задаем параметр equal_var=False, так как распредления в выборке не соовтетствуют нормальным
results = st.ttest_ind(df_product_1, df_product_2, equal_var=False)


print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

- Значение p-value показывает: вероятность случайно получить такое или большее различие мала. Таким образом, вероятность получить одинакове средние возрасты в данных выборках невелика. Делаем вывод, что средний возраст клиентов, которые пользуются двумя продуктами банка и которые пользуются одним, различается. 

### Проверим гипотезу от равенстве среднего крединого рейтинга у клиентов, которые пользуются двумя продуктами банка, и теми, которые пользуются тремя продуктами банка.

**Сформулируем гипотезы:**

Гипотеза H0: Средний кредитный рейтинг клиентов, которые пользуются двумя продуктами банка и которые пользуются тремя, одинаковый. 

Альтернативная гипотеза H1: ССредний кредитный рейтинг клиентов, которые пользуются двумя продуктами банка и которые пользуются тремя, различается. 

```python
# создадим выборки для клиентов с 2 и 3 продуктами
df_product_2 = df[df['products'] == 2]['score']
df_product_3 = df[df['products'] == 3]['score']
```

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

# задаем параметр equal_var=False, так как распредления в выборке не соовтетствуют нормальным
results = st.ttest_ind(df_product_2, df_product_3, equal_var=False)


print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

- Значение p-value показывает: вероятность случайно получить такое или большее различие слишком мала. Таким образом, вероятность получить одинакове средние возрасты в данных выборках практически отсутствует. Делаем вывод, что средний кредитный рейтинг клиентов, которые пользуются двумя продуктами банка и которые пользуются тремя, значительно различается.
